```python
INGESTION_CONFIG = [
    {
        "source": "source_erp",
        "path": "/Volumes/workspace/bronze_layer/source_system/source_erp/CUST_AZ12.csv",
        "table": "erp_cust_az12"
    },
    {
        "source": "source_erp",
        "path": "/Volumes/workspace/bronze_layer/source_system/source_erp/LOC_A101.csv",
        "table": "erp_loc_a101"
    },
    {
        "source": "source_erp",
        "path": "/Volumes/workspace/bronze_layer/source_system/source_erp/PX_CAT_G1V2.csv",
        "table": "erp_px_cat_g1v2"
    }
]
```

```python

%python

for item in INGESTION_CONFIG:
    print(f"Ingesting {item['source']} → workspace.bronze_layer.{item['table']}")

    df = (
        spark.read
             .option("header", "true")
             .option("inferSchema", "true")
             .csv(item["path"])
    )

    (
        df.write
          .mode("overwrite")
          .format("delta")
          .saveAsTable(f"workspace.bronze_layer.{item['table']}")
    )
```
